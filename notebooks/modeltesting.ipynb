```python
import os
import torch
import torchvision
import tarfile
import torch.nn as nn
import numpy as np
import torch.nn.functional as F
from torchvision.datasets.utils import download_url
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader
import torchvision.transforms as transforms
from torch.utils.data import random_split
from torchvision.utils import make_grid
import matplotlib
from torch.utils.data import Dataset, DataLoader
import matplotlib.pyplot as plt
import glob
import torch.optim as optim
import time
from torchvision import transforms as T, models

import torchvision
import torchvision.transforms as transforms
from torch.autograd import Variable
from PIL import Image, ImageOps, ImageEnhance

import cv2
%matplotlib inline

d = "cpu"
if torch.backends.mps.is_available() and torch.backends.mps.is_built():
    d = "mps"
elif torch.cuda.is_available():
    d = "cuda:0"
device = torch.device(d)

print("Device: {}".format(device))

sat_map ={0: 'Daisy',
 1:'Lavender',
 2:'Lily',
 3:'Rose',
 4:'Sunflower'}
# rsicb_map={0:"cloudy",1:"desert",2:"green_area",3:"water"}
model = torch.load("/kaggle/input/model-5-resnet/resnet5.pth")
model.eval()
transform = transforms.Compose([transforms.ToTensor()])

img = Image.open('/kaggle/input/sampledatasetfinal/Sample Dataset/Lavender (26).jpeg')  # Load image as PIL.Image
x = transform(img)  # Preprocess image
x = x.unsqueeze(0)  # Add batch dimension
x=x.to(device)
output = model(x)  # Forward pass
pred = torch.argmax(output,1)  # Get predicted class if multi-class classification
print(output)
pred_label = sat_map[pred.item()]

# pred = output.detach().cpu().tolist()[0]


print('Image predicted as ', sat_map[pred.item()])
```
